```python
#CORRELATION ANALYSIS OF COVID_19 METRICES
```

```python
covid_metrics= df[["total_confirmed","total_deaths","total_recovered","active_cases","total_tests"]]
```

```python
correlation_matrix= covid_metrics.corr()
correlation_matrix
```

```python
plt.figure(figsize =(10,6))
sns.heatmap(correlation_matrix, annot=True,cmap="coolwarm", linewidths=0.5)
plt.title("correlation matrix of covid 19 metrics")
plt.show()
```

```python
#Key Observations from the Heatmap:
#Total Confirmed vs. Total Deaths (0.91): Strong positive correlation, meaning that as confirmed cases rise, deaths also increase.

#Total Confirmed vs. Total Recovered (0.99): Very strong positive correlation, indicating that recoveries increase proportionally with confirmed cases.

#Total Deaths vs. Total Recovered (0.91): High correlation, showing that in areas with more cases, both recoveries and deaths are high.

#Total Confirmed vs. Total Tests (0.88): Indicates that more tests generally lead to more confirmed cases.

#Active Cases vs. Total Confirmed (0.72): Moderate correlation, suggesting that a portion of confirmed cases remain active.

#Active Cases vs. Total Deaths (0.58): A weaker positive correlation, meaning active cases do not always result in deaths.

#Total Tests vs. Total Deaths (0.78): Moderate correlation, showing that increased testing also leads to more recorded deaths (likely due to more detection)0.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
features= ['total_confirmed',
             'total_deaths',
             'total_recovered', 
            'active_cases',
            'serious_or_critical',
            'total_cases_per_1m_population',
            'total_deaths_per_1m_population',
            'total_tests',
            'total_tests_per_1m_population',
            'population']
X= df[features]
X
```

```python
#handle the missing value
imputer= SimpleImputer(strategy='mean')
imputer
```

```python
X_imputed= imputer.fit_transform(X)
X_imputed
```

```python
#standardize the features
scaler= StandardScaler()
```

```python
X_scaled= scaler.fit_transform(X_imputed)
X_scaled
```

```python
#PCA for dimensonality reduction
pca= PCA(n_components=2)
```

```python
X_pca= pca.fit_transform(X_scaled)
X_pca
```

```python
#KMeans clustering
kmeans= KMeans(n_clusters=3, random_state=42)
kmeans
```

```python
clusters= kmeans.fit_predict(X_pca)
```

```python
df['Cluster']= clusters
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=X_pca[:,0], y= X_pca[:,1], hue= clusters, palette='Set2')(title)
plt.title('COUNTRY CLUSTERS BASED ON COVID 19 PATTERNS')
plt.xlabel('PCA component1')
plt.ylabel('PCA component2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show
```
